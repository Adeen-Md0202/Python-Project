```python
# 

table = pd.pivot_table(data,
                          values = 'Sales',
                          index= 'State', 
                          aggfunc='sum')
```

```python
table = pd.pivot_table(
    data= data,
    columns= 'Segment',
    values = ['Sales','Profit'],
    index = ['Category','Sub-Category'],
    aggfunc=['sum','count','mean'],
    margins=True,
    margins_name='Total'
)
```

```python
table
```

```python
file = table.to_csv('Pivot_Data.csv')
```

```python
# Visualization in Python

# 11. Create a Barplot for Sub Category with Sales

plt.figure(figsize=(10,8))
sns.barplot(x = Sales_Sub_Category.index, y = Sales_Sub_Category.values,palette='deep')
plt.xticks(rotation = 90)
plt.tight_layout()
```

```python
# Pie Chart for Region with Profit

Profit_with_Region = data.groupby('Region')['Profit'].sum()
Profit_with_Region.plot(kind='pie')
plt.title('Region wise Profit in Superstore Data')
```

```python
# Distance Plot for Discount

sns.histplot(data['Quantity'],bins= 5)
```

```python
# Line Chart
# data['Month'] = data.index.strftime('%b')
data['Quarter'] = data.index.strftime('%Y') +'Q' + ((data.index.month - 1) // 3 + 1).astype(str)
```
